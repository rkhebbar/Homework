- According to the data below, two factors which impact student performance is school size and school type.
- Small and medium size schools outperformed larger ones with higher average math and reading scores as well as a higher overall passing rate
- In terms of school type, charter schools seem to do better across all metrics. Additionally, all 5 of the bottom performing schools are district schools where as the top 5 performers are charter schools 
- While there are glaring differences in terms of size and type, increases in spending per student does not guarentee higher scores or passing rates. In fact, the data seems to show the opposite. The higher the spending ranges the lower the performance across all metrics.  


Merged Data

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data, encoding="utf-8")
student_df = pd.read_csv(student_data, encoding="utf-8")


# Combine the data into a single dataset
merged_data_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
merged_data_df.head()
```

District Summary

```python
school_count = len(school_df["School ID"].unique())

student_count = len(student_df["Student ID"].unique())

total_budget= school_df['budget'].sum()

avg_math = student_df["math_score"].mean()

avg_reading = student_df["reading_score"].mean()

overall_pass = (avg_reading + avg_math)/2

num_pass_reading = student_df.loc[student_df['reading_score'] >= 70]['reading_score'].count()

perc_pass_reading = (num_pass_reading/student_count)* 100

num_pass_math = student_df.loc[student_df['math_score'] >= 70]['math_score'].count()

perc_pass_math = (num_pass_math/student_count)* 100


district_summary_df = pd.DataFrame({"Total Schools": [school_count],
                                    "Total Students": [student_count],
                                    "Total Budget": [total_budget], 
                                    "Average Math Score": [avg_math],
                                    "Average Reading Score": [avg_reading], 
                                    "% Passing Math": [perc_pass_math], 
                                    "% Passing Reading": [perc_pass_reading], 
                                    "% Overall Passing Rate": [overall_pass]})
district_summary_df.head()
```

Top Performers

```python
grouped_school = merged_data_df.set_index('school_name').groupby(['school_name'])

school_types = school_df.set_index('school_name')['type']

student_per_school = grouped_school['Student ID'].count()

school_budget = school_df.set_index('school_name')['budget']

student_budget = school_df.set_index('school_name')['budget']/school_df.set_index('school_name')['size']

avg_math = grouped_school['math_score'].mean()

avg_reading = grouped_school['reading_score'].mean()

pass_math = merged_data_df[merged_data_df['math_score'] >= 70].groupby('school_name')['Student ID'].count()/student_per_school*100

pass_read = merged_data_df[merged_data_df['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/student_per_school*100

overall = merged_data_df[(merged_data_df['reading_score'] >= 70) & (merged_data_df['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/student_per_school*100

school_summary_df = pd.DataFrame({
    "School Type": school_types,
    "Total Students": student_per_school,
    "Per Student Budget": student_budget,
    "Total School Budget": school_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_reading,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall})
school_summary_df.head(15)

top_performers = school_summary_df.sort_values("Overall Passing Rate", ascending = False)
top_performers.head()
```

Bottom Performers

Scores By Spending

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
merged_data_df['spending_bins'] = pd.cut(merged_data_df['budget']/merged_data_df['size'], spending_bins, labels = group_names)

spending_group = merged_data_df.groupby("spending_bins")


avg_math = spending_group['math_score'].mean()

avg_read = spending_group['reading_score'].mean()


pass_math = merged_data_df[merged_data_df['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/spending_group['Student ID'].count()
pass_read = merged_data_df[merged_data_df['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/spending_group['Student ID'].count()
overall = merged_data_df[(merged_data_df['reading_score'] >= 70) & (merged_data_df['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/spending_group['Student ID'].count()

score_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall})
            

score_by_spend = score_by_spend[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"]]

score_by_spend.index.name = "Per Student Budget"
score_by_spend.head()

score_by_spend = score_by_spend.reindex(group_names)

score_by_spend.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              'Overall Passing Rate': '{:.1%}'})


```
