```python
import pandas as pd
import numpy as np
```

```python
BD_0718 = pd.read_csv("Raw/JC-201807-citibike-tripdata.csv")
BD_0818 = pd.read_csv("Raw/JC-201808-citibike-tripdata.csv")
BD_0918 = pd.read_csv("Raw/JC-201809-citibike-tripdata.csv")
BD_1018 = pd.read_csv("Raw/JC-201810-citibike-tripdata.csv")
BD_1118 = pd.read_csv("Raw/JC-201811-citibike-tripdata.csv")
BD_1218 = pd.read_csv("Raw/JC-201812-citibike-tripdata.csv")
BD_0119 = pd.read_csv("Raw/JC-201901-citibike-tripdata.csv")
BD_0219 = pd.read_csv("Raw/JC-201902-citibike-tripdata.csv")
BD_0319 = pd.read_csv("Raw/JC-201903-citibike-tripdata.csv")
BD_0419 = pd.read_csv("Raw/JC-201904-citibike-tripdata.csv")
BD_0519 = pd.read_csv("Raw/JC-201905-citibike-tripdata.csv")
BD_0619 = pd.read_csv("Raw/JC-201906-citibike-tripdata.csv")
```

```python
frames = [BD_0718,BD_0818,BD_0918,BD_1018,BD_1118,BD_1218,BD_0119,BD_0219,BD_0319,BD_0419,BD_0519,BD_0619]
```

```python
BD_Data = pd.concat(frames)
```

```python
BD_Data['Start Date'], BD_Data['Start Time'] = BD_Data['starttime'].str.split(' ', 1).str
BD_Data['Stop Date'], BD_Data['Stop Time'] = BD_Data['stoptime'].str.split(' ', 1).str
```

```python
BD_Data.head()
```

```python
BD_Data['Start Time'], BD_Data['Junk'] = BD_Data['Start Time'].str.split('.').str
```

```python
BD_Data.head()
```

```python
BD_Data['Stop Time'], BD_Data['Junk2'] = BD_Data['Stop Time'].str.split('.').str
```

```python
BD_Data.head()
```

```python
BD_Data.drop(columns=['starttime', 'stoptime', 'Junk', 'Junk2'], inplace = True)
```

```python
BD_Data.to_csv('Resources/BD_Set.csv')
```
